# Formula 1 Driver Comparison

Formula 1 is a global motorsport where 10 teams compete, each fielding 2 cars and 2 drivers, to race for the world championship across a series of events, called Grand Prixs, held on tracks around the world. Each team’s drivers aim to finish races in the highest positions possible, earning points for themselves and their team. At the end of the season, the driver and team with the most points are crowned world champions. Each F1 team designs and builds its own unique car given the same strict technical rules and set budget. Each team, however, uses the same exact car, replicas of each other, for both drivers. This in-turn makes the main difference between where each team's drivers finish in a race come down to the drivers themselves. When both drivers have access to the same equipment, any gap in performance is most likely due to their individual skill. Skill gaps in F1 are apparent most during corners. Due to both drivers having the same car, and weighing basically the same physically, advantages or differences in track straights are rare. This leaves us with corners; differences between car path, brake points, brake distances, throttle pickups, and how a driver combines these together throughout a race's corners. We want to take a deep dive into this idea and see if this theory holds when looking at the actual data. Are differences between drivers of the same team created in the corners, and if so: how would one use this knowledge to improve and optimize their race times?

To attempt this we need to ensure all drivers are given the same fair chance. As a result, we will be looking into each driver's fastest lap during the qualifying round of a race. Before every race there is a qualifying round. This round determines the starting order of the drivers on the grid. During this round, drivers are given the most optimal conditions possible to ensure fairness across the board. Cars are set up for maximum speed, given just enough fuel for the laps, fresh tires, and the least possible weight. Drivers are allowed, encouraged even, to push to the limit without worry of the factors that would otherwise be crucial during the race, such as tire wear, race strategy, or fuel management. Therefore, hypothetically, this makes a driver's fastest lap in the qualifying round a matter of just their pure driving ability.

To start attempting to answer our question we first need to find and gather clean data on F1 races and drivers.

## Data

For this project we will be using the FastF1 API. This Python library gathers data directly form the official F1 timing feeds and APIs. The data includes lap times, car telemetry, weather, results, and more. FastF1 uses pandas DataFrames asthe primary way of delivering data. Below is an example of data loaded from the API.

```python
import fastf1

#load a session
session = fastf1.get_session(2022, 'Monaco Grand Prix', 'Q')
session.load()

#choose a driver's fastes lap
fastest_lap = session.laps.pick_drivers('LEC').pick_fastest()

#gather telemetry data
tel = fastest_lap.get_car_data().add_distance()

#display data
print(tel[['Brake', 'Speed', 'nGear', 'Throttle', 'Distance']].head())


```
